```python
import math
import numpy as np
from qiskit.visualization import array_to_latex, plot_histogram
from qiskit.quantum_info import Statevector, Operator

psi1 = Statevector([math.cos(math.pi/8), math.sin(math.pi/8)])
psi2 = Statevector([math.cos(5*math.pi/8), math.sin(5*math.pi/8)])

display(array_to_latex(psi1))
display(array_to_latex(psi2))

H = Operator(psi1) - Operator(psi2)
display(array_to_latex(H))
```

```python
import matplotlib.pyplot as plt

# Single qubit phase estimation
# p_0 = cos^2(pi. theta)
# p_1 = sin^2(pi. theta)

theta = np.linspace(0, 1, 100)
p_0 = []
p_1 = []
for i in theta:
    p_0.append(math.cos(math.pi * i) ** 2)
    p_1.append(math.sin(math.pi * i) ** 2)
#print(theta)
#print(p_0)
#print(p_1)

y_values = [p_0, p_1]  # List of functions
labels = [r"$0$", r"$1$"]      # Corresponding labels
colors = ["b", "r"]                # Colors for the plots

for i, y in enumerate(y_values):
    plt.plot(theta, y, label = labels[i], color = colors[i])

plt.xlabel(r"$\theta$")
plt.ylabel("Probalibity")
plt.title("Plot of $p_0$ and $p_1$ (single-qubit phase estimation for U)")
plt.legend()

plt.show()
```

```python
from math import pi, cos, sin
from qiskit import QuantumCircuit

theta = 0.6  # Can be changed to any value between 0 and 1

qc = QuantumCircuit(2, 1)

# Prepare eigenvector, which is the |1> state
qc.x(1)
qc.barrier()

qc.h(0)
qc.cp(2 * pi * theta, 0, 1)
qc.h(0)
qc.barrier()

# Perform the final measurement
qc.measure(0, 0)


qc.draw('mpl')
```

```python
from qiskit.primitives import Sampler

display(Sampler().run(qc).result().quasi_dists[0])
```

```python
display({
    0: cos(pi * theta) ** 2, 
    1: sin(pi * theta) ** 2
 })
```

```python
from qiskit_aer import AerSimulator

result = AerSimulator().run(qc, shots = 1000).result()
statistics = result.get_counts()
display(plot_histogram(statistics))
```

```python
import matplotlib.pyplot as plt

# Single qubit phase estimation
# p_0 = cos^2(pi. theta)
# p_1 = sin^2(pi. theta)

theta = np.linspace(0, 1, 100)
p_0 = []
p_1 = []
for i in theta:
    p_0.append(math.cos(math.pi * 2 * i) ** 2)
    p_1.append(math.sin(math.pi * 2 * i) ** 2)
#print(theta)
#print(p_0)
#print(p_1)

y_values = [p_0, p_1]  # List of functions
labels = [r"$0$", r"$1$"]      # Corresponding labels
colors = ["b", "r"]                # Colors for the plots

for i, y in enumerate(y_values):
    plt.plot(theta, y, label = labels[i], color = colors[i])

plt.xlabel(r"$\theta$")
plt.ylabel("Probalibity")
plt.title("Plot of $p_0$ and $p_1$ (single-qubit phase estimation for $U^2$)")
plt.legend()

plt.show()
```

```python
# Two-qubit phase estimation probabilities
theta = np.linspace(0, 1, 100)
p_00 = []
p_01 = []
p_10 = []
p_11 = []

for i in theta:
    x_1 = math.cos(math.pi * 2 * i)
    x_2 = math.cos(math.pi * 4 * i)
    x_3 = math.cos(math.pi * 6 * i)

    y_1 = math.sin(math.pi * 2 * i)
    y_2 = math.sin(math.pi * 4 * i)
    y_3 = math.sin(math.pi * 6 * i)

    p_00.append(((1 + x_1 + x_2 + x_3)/4) ** 2 + ((y_1 + y_2 + y_3)/4) ** 2)
    p_01.append(((1 - y_1 - x_2 + y_3)/4) ** 2 + ((x_1 - y_2 - x_3)/4) ** 2)
    p_10.append(((1 - x_1 + x_2 - x_3)/4) ** 2 + ((-y_1 + y_2 - y_3)/4) ** 2)
    p_11.append(((1 + y_1 - x_2 - y_3)/4) ** 2 + ((-x_1 - y_2 + x_3)/4) ** 2)
    
y_values = [p_00, p_01, p_10, p_11]  # List of functions
labels = [r"$00$", r"$01$", r"$10$", r"$11$"]      # Corresponding labels
colors = ["b", "r", "g", "y"]                # Colors for the plots

for i, y in enumerate(y_values):
    plt.plot(theta, y, label = labels[i], color = colors[i])

plt.xlabel(r"$\theta$")
plt.ylabel("Probalibity")
plt.title("Two-qubit phase estimation")
plt.legend()

plt.show()
```

```python
from qiskit import QuantumCircuit

theta = 0.3

qc = QuantumCircuit(3, 2)

# Define the eigen vector
qc.x(2)
qc.barrier()

# The initial Hadamard Gates
qc.h(0)
qc.h(1)
qc.barrier()

# The controlled unitary gates
qc.cp(2 * pi * theta, 0,2)
qc.cp(2 * pi * (2 * theta), 1, 2)
qc.barrier()

# The implementation of the inverse of the QFT
qc.swap(0,1)
qc.h(0)
qc.cp(- pi/2, 0,1)
qc.h(1)
qc.barrier()

# And finally the measurements
qc.measure([0, 1], [0, 1])
display(qc.draw(output="mpl"))
```

```python
result = AerSimulator().run(qc, shots = 1000).result()
statistics = result.get_counts()

display(plot_histogram(statistics))
```
