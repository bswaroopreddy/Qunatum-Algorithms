```python
import numpy as np
from qiskit.quantum_info import Statevector

# Define quantum states
u = Statevector([1, 0])  # |0⟩
v = Statevector([0, 1])  # |1⟩

# Compute outer product
outer_product = np.outer(u.data, v.data.conj())

print("Outer Product:\n", outer_product)
```

```python
from qiskit.quantum_info import DensityMatrix

# Define quantum states
u = Statevector([1, 0])  # |0⟩
v = Statevector([0, 1])  # |1⟩

# Compute the outer product |u⟩⟨v| as a DensityMatrix
outer_product_matrix = DensityMatrix(np.outer(u.data, v.data.conj()))

print("Outer Product as Density Matrix:\n", outer_product_matrix)
```

```python
# Generate Basis vectors
from qiskit.quantum_info import Statevector

def get_basis_statevector(n):
    """Generate computational basis statevectors for an n-qubit system."""
    basis_statevectors = []
    for i in range(2**n):
        # Generate basis state |i⟩
        statevector = Statevector.from_label(format(i, f'0{n}b'))
        basis_statevectors.append(statevector)
    return basis_statevectors

# Example: Basis vectors for a 3-qubit system
basis_statevectors = get_basis_statevector(3)

# Print basis statevectors
for vec in basis_statevectors:
    print(vec, "\n")
```

```python
# Phase estimation using QFT

from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.circuit.library import QFT
from qiskit.visualization import plot_histogram
from math import pi

theta = 0.7
m = 3

control = QuantumRegister(m, name = "X")
target = QuantumRegister(1, name = "Y")
output = ClassicalRegister(m, name = "Z")

qc = QuantumCircuit(control, target, output)

#Prepare the eigen vector
qc.x(target)
qc.barrier()

# Perform phase estimates
for index, qubit in enumerate(control):
    qc.h(qubit)

    for _ in range(2**index):
        qc.cp(2 * pi * theta, qubit, target)

qc.barrier()

qc.compose(QFT(m, inverse = True).decompose(), inplace = True)
qc.barrier()

# Measuremets
qc.measure(range(m), range(m))


display(qc.draw("mpl"))
```

```python
result = AerSimulator().run(qc).result()
statistics = result.get_counts()

display(plot_histogram(statistics))
```

```python
most_probable = int(max(statistics, key=statistics.get),2)
print(f"Most probable output: {most_probable}")
print(f"Estimated theta: {most_probable/2**m}")
```
