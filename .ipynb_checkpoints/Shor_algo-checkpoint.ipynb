```python
import numpy as np
from math import gcd, floor, log
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.circuit.library import UnitaryGate, QFT
from qiskit.quantum_info.operators import Operator
import random
from fractions import Fraction
from qiskit_aer import AerSimulator
```

```python
# Function to create the gate M_b for b in Z*_n
def mod_mul_gate(b, N):
    if gcd(b, N) > 1:
        print(f"Error: gcd({b}, {N}) > 1")
    else:
        n = floor(log(N-1, 2)) + 1
        U = np.full((2**n, 2**n), 0)
        for x in range(0, N): U[b * x % N][x] = 1
        for x in range(N, 2**n): U[x][x] = 1
        G = UnitaryGate(U)
        G.name = f"M_b"

        return G
```

```python
def order_finding_circuit(a, N):
    if gcd(a, N) > 1:
        print(f"Error: gcd({a}, {N}) > 1")
    else:
        n = floor(log(N-1, 2)) + 1
        m = 2*n

        control = QuantumRegister(m, name = "X")
        target = QuantumRegister(n, name = "Y")
        output = ClassicalRegister(m, name = "Z")

        circuit = QuantumCircuit(control, target, output)
        # Initialize the target register to the state |1>
        circuit.x(m)

        # Add the Hadamard gates and controlled versions of the multiplication gates
        for i, qubit in enumerate(control):
            circuit.h(i)
            b = pow(2**i, N)
            circuit.compose(
                mod_mul_gate(b, N).control(),
                qubits = [qubit] + list(target),
                inplace = True
            )

        # Apply the inverse QFT to the control register
        circuit.compose(
            QFT(m, inverse = True),
            control,
            inplace = True
        )

        # output measurement
        circuit.measure(control, output)

        return circuit
```
